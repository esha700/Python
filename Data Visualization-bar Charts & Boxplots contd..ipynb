```python
#PART B: Using the other dataframe
df = pd.read_csv('LaptopSales.csv', parse_dates=["Date"], index_col="Date")
df.head()
```

```python
df.columns = [s.strip().replace(' ', '_') for s in df.columns] 
list(df.columns)
```

```python
#PART B- Monthly Time Series
df.Retail_Price.resample("M").mean().plot()
```

```python
#PART B- WEEKLY TIME SERIES 
df.Retail_Price.resample("W").mean().plot()
```

```python
#PART B-DAILY TIME SERIES
df.Retail_Price.resample("D").mean().plot()
```

```python
#PART_B:Quarterly Time Series
df.Retail_Price.resample("Q").mean().plot()
```

```python
#PART_B- daily frequency without resampling is very granular.SO, it is hard to see 
df.Retail_Price.plot()
```

```python
#PART-C: Checking the consistency of prices across various stores
#Answer: We see that prices are conistent between $450-$525 across various stores with outliers in case of all the store locations
df.groupby('Store_Postcode').mean().Retail_Price
```

```python
ax = df.boxplot(column='Retail_Price', by='Store_Postcode')
ax.set_ylabel('Price')
ax.set_xlabel('Store')
plt.suptitle('')  # Suppress the titles
plt.title('')
plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right', fontsize='medium')

plt.show()
```

```python
#PART-D: Change in price as per configuration- The price has increased as configurations have changed
df.groupby('Configuration').mean().Retail_Price
```

```python
ax = df.boxplot(column='Retail_Price', by ='Configuration',figsize=[200,3])
ax.set_ylabel('Price')
ax.set_xlabel('Configuration')
plt.suptitle('')  # Suppress the titles
plt.title('')
plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right', fontsize='small')
plt.show()
#double click and scroll to view the result
```
