```python
df = pd.read_csv('ApplianceShipments.csv',parse_dates=["Quarter"], index_col="Quarter")
```

```python
#PART-B:Plotting 3500-5000 limit on Y axis - It shows the shipments are peaking usually around second quarter each year and 
#then falling to their lowest levels in the last quarter of each year.
plt.xlabel("Quarter")
plt.ylabel("Shipment in million dollars")
plt.title("Quarterly Shipment of Household Appliances")
df.Shipments.plot()
plt.ylim(3500,5000)
```
