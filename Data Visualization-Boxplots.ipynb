```python
#Laptop Sales at a London Computer Chain: Bar Charts and Boxplots. The file LaptopSalesJanuary2008.csv contains data for all 
#sales of laptops at a computer chain in London in January 2008. This is a subset of the full dataset that includes
#data for the entire year.
#a. Create a bar chart, showing the average retail price by store.Which store has the highest average? Which has the lowest?
#b. To better compare retail prices across stores, create side-byside boxplots of retail price by store
```

```python
#PART A
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('LaptopSalesJanuary2008.csv')
```

```python
df.rename (columns={'Customer Postcode': 'Customer_Postcode'}) 
list(df.columns)
```

```python
df.columns = [s.strip().replace(' ', '_') for s in df.columns] 
list(df.columns)
```

```python
df.groupby('Store_Postcode').mean().Retail_Price
```

```python
ax = df.groupby('Store_Postcode').mean().Retail_Price.plot(kind='bar', title="Average Retail Price by Store")
ax.set_ylabel('Avg. Retail Price')
plt.ylim(470, 500)
```

```python
#PART-A
#Store with highest Average Retail Price = N17 6QA
#Store with lowest Average Retail Price = W4 3PH
```

```python
#PART B- Side by Side Boxplots
```

```python
Store = [s.strip().replace(' ', '_') for s in df.Store_Postcode] 
list(Store)
```

```python
df.groupby('Store_Postcode').mean().Retail_Price
```

```python
ax = df.boxplot(column='Retail_Price', by='Store_Postcode')
ax.set_ylabel('Price')
ax.set_xlabel('Store')
plt.suptitle('')  # Suppress the titles
plt.title('')
plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right', fontsize='medium')

plt.show()
```

```python
#Answer(PART B): We can see that stores with postal code N17_6QA has a higher average price and quartiles as compared to the store in W4_3PH 
#It is interesting to see that maximum prices are almost similar in both store locations 
#Not many outliers in N17_6QA as compared to W4_3PH
```
