```python
%matplotlib inline
from pathlib import Path

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge, LassoCV, BayesianRidge
import statsmodels.formula.api as sm
import matplotlib.pylab as plt

from dmba import regressionSummary, exhaustive_search
from dmba import backward_elimination, forward_selection, stepwise_selection
from dmba import adjusted_r2_score, AIC_score, BIC_score
import warnings
warnings.filterwarnings('ignore')
import statsmodels.api as sm
import statsmodels.formula.api as sm
```

```python
car_df= pd.read_csv('ToyotaCorolla.csv')
```

```python
car_df.dtypes
```

```python
predictors = ['Age_08_04', 'KM', 'Fuel_Type', 'HP', 'Automatic', 'Airco', 
              'Doors', 'Quarterly_Tax', 'Automatic_airco','Mfr_Guarantee','Guarantee_Period','CD_Player','Powered_Windows','Sport_Model','Tow_Bar']
outcome = 'Price'
```

```python
#Creating dummies
X = pd.get_dummies(car_df[predictors], drop_first=True)
Y = car_df[outcome]
```

```python
#Splitting the data into training, test and validation
train_X, valid_X, train_Y, valid_Y = train_test_split(X, Y, test_size=0.5, random_state=1)
test_X, valix_X, test_Y, valid_Y = train_test_split(valid_X, valid_Y, test_size=0.6, random_state=1)
print('Training :', train_X.shape)
print('Validation :', valid_X.shape)
print('Test :',test_X.shape)
```

```python
car_lm = LinearRegression()
car_lm.fit(train_X, train_Y)
```

```python
# print coefficients
print('intercept ', car_lm.intercept_)
print(pd.DataFrame({'Predictor': X.columns, 'coefficient': car_lm.coef_}))

# print performance measures
regressionSummary(train_Y, car_lm.predict(train_X))
```

```python
pred_Y = car_lm.predict(train_X)

print('adjusted r2 : ', adjusted_r2_score(train_Y, pred_Y, car_lm))
print('AIC : ', AIC_score(train_Y, pred_Y, car_lm))
print('BIC : ', BIC_score(train_Y, pred_Y, car_lm))
```

```python
#PART-B
#Splitting the dataset into training and validation as 60-40
train_X, valid_X, train_Y, valid_Y = train_test_split(X, Y, test_size=0.4, random_state=1)
```

```python
car_lm = LinearRegression()
car_lm.fit(train_X, train_Y)
print('intercept ', car_lm.intercept_)
print(pd.DataFrame({'Predictor': X.columns, 'coefficient': car_lm.coef_}))
```

```python
pred_Y = car_lm.predict(train_X)

print('adjusted r2 : ', adjusted_r2_score(train_Y, pred_Y, car_lm))
print('AIC : ', AIC_score(train_Y, pred_Y, car_lm))
print('BIC : ', BIC_score(train_Y, pred_Y, car_lm))
```

```python
# Use predict() to make predictions on a new set
car_lm_pred = car_lm.predict(valid_X)

result = pd.DataFrame({'Predicted': car_lm_pred, 'Actual': valid_Y,
                       'Residual': valid_Y - car_lm_pred})
print(result.head(20))

# Compute common accuracy measures
regressionSummary(valid_Y, car_lm_pred)

```

```python
#determining residuals
car_lm_pred = car_lm.predict(valid_X)
all_residuals = valid_Y - car_lm_pred

# Determine the percentage of datapoints with a residual in [-1500, 1500] = approx. 83\%
print(len(all_residuals[(all_residuals > -1500) & (all_residuals < 1500)]) / len(all_residuals))

ax = pd.DataFrame({'Residuals': all_residuals}).hist(bins=25)
plt.tight_layout()
plt.savefig("myexample.png")
```
